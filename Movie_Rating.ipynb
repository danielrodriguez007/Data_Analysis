```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
movies = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/movies.dat', delimiter = '::')
movies.head()
```

```python
movies.columns = ["ID", "Title","Genre"]
movies.head()
```

```python
ratings = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/ratings.dat', delimiter='::')
ratings.head()
```

```python
ratings.columns = ["User", "ID", "Ratings", "Timestamp"]
ratings.head()
```

```python
data = pd.merge(movies, ratings, on=["ID", "ID"])
data.head()
```

```python
ratings = data['Ratings'].value_counts()
numbers = ratings.index
quantity = ratings.values
import plotly.express as px 
fig = px.pie(data, values=quantity, names=numbers)
fig.show()
```
