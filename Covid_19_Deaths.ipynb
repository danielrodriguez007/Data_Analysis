```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/COVID19 data for overall INDIA.csv')
data.head()
```

```python
data.isnull().sum()
```

```python
data = data.drop('Date', axis=1)
```

```python
figure = px.bar(data, x='Date_YMD', y = 'Daily Deceased')
figure.show()
```

```python
cases = data['Daily Confirmed'].sum()
deceased = data['Daily Deceased'].sum()

labels = ["Confirmed", "Deceased"]
values = [cases, deceased]
figure = px.pie(data, values=values, names = labels,
                title="Daily Confirmed Cases vs Daily Deaths", hole=0.5)
figure.show()
```

```python
date_rate = (data['Daily Deceased'].sum()/data['Daily Confirmed'].sum())*100
print(np.round(date_rate,2),"%")
```

```python
figure = px.bar(data,x='Date_YMD', y='Daily Deceased')
figure.show()
```
