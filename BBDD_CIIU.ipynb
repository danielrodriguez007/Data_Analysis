```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import plotly.express as px 
import plotly.graph_objects as go 
plt.style.use('fivethirtyeight')
```

```python
data = pd.read_json('https://www.datos.gov.co/resource/tnnf-6twt.json')
data.head(5)
```

```python
data.columns
```

```python
data.isnull().sum()
```

```python
data2 = data[['matricula', 'organizacion', 'categoria', 'estado_matricula',
       'est_datos', 'razon_social', 'clase_id', 'identificacion', 'nit',
       'nacionalidad']]
data2.head()       
```

```python
data2.isnull().sum()
```

```python
data2['nacionalidad'] = data2['nacionalidad'].fillna('DESCONOCIDA')
```

```python
data2['nacionalidad'] = data2['nacionalidad'].map({'COLOMBIANA':'COLOMBIANA','COLOMBIANO':'COLOMBIANA','COLOMBIANO/A':'COLOMBIANA','DESCONOCIDA':'DESCONOCIDA'})
```

```python
data2.info()
```

```python
pivot1 = pd.pivot_table(data2, values = 'matricula', index = 'nacionalidad', aggfunc='count')
pivot1
```

```python
pivot2 = pd.pivot_table(data2, values='matricula',index=['nacionalidad','organizacion'], columns='estado_matricula',aggfunc='count')
pivot2
```
