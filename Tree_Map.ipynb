```python
import seaborn as sns
import matplotlib.pyplot as plt
import squarify
sns.set()
```

```python
data = sns.load_dataset('titanic')
print(data.shape)
data.head()
```

```python
n_data = data.groupby('embark_town')[['survived']].sum()
n_data
```

```python

plt.figure(figsize=(10,8))
sizes = n_data.values
label = n_data.index
squarify.plot(sizes=sizes, label=label,)
plt.axis('off')
plt.title('First Tree Map', fontsize=20)
```

```python
data.head()
```

```python
c_df = data.groupby('class')[['survived']].count()
c_df
```

```python
sizes = c_df.values
label = c_df.index 

plt.figure(figsize=(10,8))
squarify.plot(sizes=sizes, label=label, color = sns.color_palette("coolwarm")) 
plt.title('Class of the Titanic', fontsize=25)
plt.axis('off')
```

```python
data2 = sns.load_dataset('mpg')
data2.head()
```

```python
c_cars = data2.groupby('origin')[['cylinders']].count()
c_cars
```

```python
sizes = c_cars.values 
label = c_cars.index 

plt.figure(figsize=(10,8))
squarify.plot(sizes=sizes, label=label,color = sns.color_palette("coolwarm"))
plt.title('Cars Origin')
```

```python
cy_cars = data2.groupby('cylinders')[['origin']].count()
cy_cars
```

```python
sizes = cy_cars.values 
label = cy_cars.index 

plt.figure(figsize=(10,8))
squarify.plot(sizes=sizes, label=label, color = sns.color_palette("cool_r")) 
plt.title('Tree Map Car Cylinders', fontsize=25)
plt.axis('off')
```
