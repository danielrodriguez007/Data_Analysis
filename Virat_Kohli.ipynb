```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go 
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/amankharwal/Website-data/master/Virat_Kohli.csv')
data.head()
```

```python
data.isnull().sum()
```

```python
data['Runs'].sum()
```

```python
np.around(data.Runs.mean(),2)
```

```python
matches = data.index
figure = px.line(data_frame=data,x=matches, y=data.Runs,
                 title='Runs Scored by Virat Kohli Between 18-Aug-08 - 22-Jan-17')
figure.show()
```

```python
data['Pos'].unique()
```

```python
data['Pos'] = data['Pos'].map({3.0: "Batting At 3", 4.0: "Batting At 4", 2.0: "Batting At 2", 
                               1.0: "Batting At 1", 7.0:"Batting At 7", 5.0:"Batting At 5", 
                               6.0: "batting At 6"})

```

```python
POS = data.groupby(['Pos'])['Pos'].count()
POS
```

```python
labels=POS.index
x=POS.values
explode=[0,0,0.1,0,0,0,0]
colors = ['#ff6666', '#ffcc99', '#99ff99', '#66b3ff','#c2c2f0','#ffb3e6','#c2c2f0']

plt.figure(figsize=(10,10))
plt.pie(x=x,labels=labels, shadow=True,explode=explode, autopct='%1.1f%%', colors=colors)
plt.title('Number of Matches At Different Batting Positions', fontsize=15)
plt.legend()
plt.show()
```

```python
RUN = data.groupby(['Pos'])['Runs'].sum()
RUN
```

```python
labels = RUN.index 
x = RUN.values 
colors = ['#ff6666', '#ffcc99', '#99ff99', '#66b3ff','#c2c2f0','#ffb3e6','#c2c2f0']
explode=[0,0,0.1,0,0,0,0]


plt.figure(figsize=(10,10))
plt.pie(x=x, labels=labels, shadow=True, colors=colors, explode=explode, autopct='%1.1f%%' )
plt.legend()
plt.title('Runs By Virat Kohli At Different Batting Positions', fontsize=15)
plt.show()
```

```python
centuries = data.query("Runs>=100")

figure = px.bar(centuries, x=centuries['Inns'], y=centuries['Runs'],
                 color=centuries['Runs'], title=("Centuries By Virat Kohli in First Innings Vs. Second Innings"))
figure.show()
```

```python
figure = px.bar(data,x=data['Opposition'], y=data['Runs'], color=data['Runs'],
                title="Most Runs Against Teams")
figure.show()
```

```python
figure = px.bar(centuries, x=centuries['Opposition'], y=centuries['Runs'],
                color=centuries['Runs'], title="Most Centuries Against Teams")
figure.show()
```

```python
figure = px.scatter(data_frame=data, x='Runs', y='4s',
                    size='SR', trendline='ols', title="Relationship Between Runs Scored and Fours")
figure.show()
```

```python
def find_duplicates(x):
  lenght = len(x)
  duplicates=[]
  for i in range(lenght):
    n  = i + 1
    for a in range(n, lenght):
      if x[i] == x[a] and x[i] not in duplicates:
        duplicates.append(x[i])
  return duplicates

names = ["Didier", "Akanksha", "Didier", "Devyansh", 
         "Aman", "Diksha", "Didier", "Devyansh"]

print(find_duplicates(names))                 
```
