```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
sns.set()


data = [5.,25., 50., 20.]
plt.bar(range(len(data)), data)
plt.show()
```

```python
data = [5.,25., 50., 20.]
plt.bar(range(len(data)), data, width=1.)
plt.show()
```

```python
data = [5.,25.,50.,20.]
plt.barh(range(len(data)),data)
plt.show()
```

```python
import numpy as np
data = [[4., 27., 50., 20.],
        [6., 25., 50., 17.],
        [8., 23., 50., 19.]]
x = np.arange(4) 

plt.bar(x + 0.00, data[0], color='b', width=0.25)
plt.bar(x + 0.25, data[1], color='r', width=0.25)
plt.bar(x + 0.50, data[2], color='g', width=0.25)
plt.show()
```

```python
a = [4., 27., 50., 20.] 
b = [6., 25., 50., 17.]

plt.bar(x,a,color = 'b')
plt.bar(x, b, color = 'r', bottom=a)
plt.show()
```

```python
x = [24, 26, 28, 30, 32, 34]
y = [15435, 29765, 33356, 47987, 50987, 67895]

plt.figure(figsize=(15,10))
plt.bar(x, y, color = "#444444", label='Dev', width=1)

py_y = [25435, 39765, 43356, 57987, 60987, 77895] 
plt.plot(x, py_y, color="#e5ae38", label='Python',linewidth=5, markersize=12 )

js_y = [15435, 29765, 33356, 47987, 50987, 67895] 
plt.plot(x, js_y, color="#008fd5", label='Js', linewidth=5, markersize=12)

plt.legend()

plt.title("Salary by Age")
plt.xlabel("Age")
plt.ylabel("Salary (USD)")
plt.show()

```

```python
## Top 10 Bar charts of countries by population

x = ['United States', 'Indonesia', 'Pakistan', 'Brazil', '	Nigeria', 'Bangladesh', 'Russia', 'Mexico', 'Japan']
y = [331002651, 273523615, 220892340, 212559417, 206139589, 164689383, 145934462, 128932753, 126476461]
y2 = [231002651, 173523615, 120892340, 112559417, 106139589, 64689383, 45934462, 28932753, 26476461]

plt.figure(figsize=(15,10))
plt.plot(x, y, label = '2022',linewidth = 3)

plt.plot(x, y2, label= '1950', color='g', linewidth = 3)

plt.legend()
plt.title('Top 10 of countries by population 1950 2022')
plt.xlabel('Countrys')
plt.ylabel('Population(Millions)')

plt.show()

```

```python
#Bar chart of the most videogame saled 

x = ["MINECRAFT", "GTA", "TETRIS", "WII SPORTS", "PUBG", "MARIO BROS", "POKEMON", "MK8", "WII FIT", "PACMAN"]
y = [238000, 150000, 100000, 82900, 70000, 58000, 47520, 45530, 438000, 40955]
y_c = [574560, 32489, 89887, 70032, 65121, 40987, 10987, 98876, 423455, 69877]
y_a = [474560, 22489, 79887, 60032, 55121, 30987, 10987, 88876, 323455, 59877]

plt.figure(figsize=(15,10))
plt.bar(x, y, label = "VideoGames", color = "#444444")
for i in range(len(x)):
  plt.text(i, y[i], y[i], ha="center")
plt.plot(x, y_c, label = "N of Players",color = "lightgreen",  linewidth =5 )
plt.plot(x, y_a, label = "N of Console",color = "lightblue",  linewidth =5 )
plt.legend()
plt.title("The most saled video games")
plt.xlabel("VideoGames")
plt.ylabel("Sales (USD)")
plt.show()

```

```python
#Horizontal barchart 

x = ["A", "B", "C", "D", "E", "F"]
y = [1, 2, 3, 4, 5, 6, ]



plt.figure(figsize=(15,10))
plt.legend()
plt.barh(x, y, label = "Test")
for i in range(len(x)):
  plt.text(i, x[i], y[i])
plt.title("Test")
plt.show()

```
