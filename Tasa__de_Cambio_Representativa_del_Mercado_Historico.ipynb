```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
sns.set()
```

```python
#import data from JSON
data = pd.read_json('https://www.datos.gov.co/resource/mcec-87by.json')
data.head()
```

```python
#Search for Null values
data.isnull().sum()
```

```python
#convert to datetime format
data['vigenciadesde'] = pd.to_datetime(data['vigenciadesde'], format = "%Y-%m-%dT%H:%M:%S.%f")
```

```python
#convert to datetime format
data['vigenciahasta'] = pd.to_datetime(data['vigenciahasta'], format = "%Y-%m-%dT%H:%M:%S.%f")
```

```python
#plot times series

import plotly.express as px

plt.figure(figsize=(15,10))
figure = px.line(data, x= data.vigenciadesde, y = "valor", title = 'Tasa _de_Cambio_Representativa_del_Mercado_Historico')
figure.show()
```
