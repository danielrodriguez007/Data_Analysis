```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
movies  = pd.read_csv('https://raw.githubusercontent.com/amankharwal/Website-data/master/MoviesOnStreamingPlatforms_updated.csv')
movies.head()
```

```python
#realized modification to the column Rotten Tomatoes and eliminated column Type
movies['Rotten Tomatoes'] = movies['Rotten Tomatoes'].str.replace("%", "").astype(float) 
movies.drop("Type", inplace=True, axis=1)
correlations = movies.corr(method = 'pearson') 
```

```python
#plotting heatmap

plt.figure(figsize=(10,8))
sns.heatmap(correlations, cmap='coolwarm')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(correlations, cmap='coolwarm', annot=True)
plt.show()
```
