# Import libraries and download data 📚

Source :https://thecleverprogrammer.com/2023/05/08/stock-market-performance-analysis-using-python/

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn
import yfinance as yf
import plotly.express as px
from datetime import datetime
plt.style.use('fivethirtyeight')
```

```python
start_date = datetime.now() - pd.DateOffset(months=3)
end_date = datetime.now()

tickers = ['AAPL','MSFT','NFLX','GOOG']
df_list=[]

for ticker in tickers:
  data = yf.download(ticker, start=start_date, end=end_date)
  df_list.append(data)

df = pd.concat(df_list, keys=tickers,names=['Ticker','Date'])
df.head()
```

```python
df = df.reset_index()
df.head()
```

# Data Analysis 👨‍💻

```python
a = df.query('Ticker=="GOOG"')
b = df.query('Ticker=="AAPL"')
c = df.query('Ticker=="MSFT"')
d = df.query('Ticker=="NFLX"')



plt.figure(figsize=(25,8))
plt.plot(a['Date'], a['Close'], label = "GOOG", color='#888888')
plt.plot(b['Date'], b['Close'], label = "APPL", color='#666666')
plt.plot(c['Date'], c['Close'], label = "MSFT",color='#00A9E0')
plt.plot(d['Date'], d['Close'], label = "NFLX",color='#59C7EB')
plt.title("Stock Market Performance for the last 3 Months")
plt.legend()
plt.ylabel("Close")
plt.xlabel("Date")
plt.show()
```

```python
plt.figure(figsize=(25,8))


plt.fill_between(d['Date'], d['Close'], label = "NFLX",color='#446DA8')
plt.fill_between(c['Date'], c['Close'], label = "MSFT",color='#008ECE')
plt.fill_between(b['Date'], b['Close'], label = "APPL",color='#888888')
plt.fill_between(a['Date'], a['Close'], label = "GOOG",color='#666666')

plt.title("Stock Market Performance for the last 3 Months")
plt.legend()
plt.ylabel("Close")
plt.xlabel("Date")
plt.show()
```
