```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
sns.set()
```

```python
#Plotting a histogram
data = np.random.multivariate_normal([0, 0], [[5, 2], [2, 2]], size = 2000)
data = pd.DataFrame(data, columns=['x', 'y'])
plt.figure(figsize=(10,8))
plt.hist(data['x'], alpha = 0.5) 
plt.hist(data['y'], alpha = 0.5)
plt.show()
```

```python
#Plotting Density graph

plt.figure(figsize=(10,8))
sns.kdeplot(data['x'], shade = True)
sns.kdeplot(data['y'], shade = True)
plt.show()
```

```python
#Visualizing both plots Histogram and Density

plt.figure(figsize=(10, 8))
sns.distplot(data['x'])
sns.displot(data['y'])
plt.show()
```

```python
data2 = np.array([[60, 61, 87, 86],[64, 65, 86, 89],[68,69,89,88], [62,63,88,85]])
df = pd.DataFrame(data2, columns=["1","2","3", "4"])
plt.figure(figsize=(10,8))
sns.kdeplot(df['1'], shade = True, label = '1')
sns.kdeplot(df['2'], shade = True, label = '2')
sns.kdeplot(df['3'], shade = True, label = '3')
sns.kdeplot(df['4'], shade = True, label = '4')
plt.legend()
plt.title("Looking how to works Density plots")
plt.xlabel("Quantity of samples")
plt.show()
print(df)
```
