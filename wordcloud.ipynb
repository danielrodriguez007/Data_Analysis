```python
from wordcloud import WordCloud, ImageColorGenerator, STOPWORDS
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Crea la mascara o forma del wordcloud
my_mask = np.array(Image.open('/content/drive/MyDrive/Colab Notebooks/Batman.png'))
stopwords = set(STOPWORDS)
```

```python
#Especificaciones del wordcloud

wc = WordCloud(background_color='black',
               mask=my_mask,
               stopwords=stopwords,)
```

```python
#Abriendo el archivo de texto
with open('/content/drive/MyDrive/Colab Notebooks/batman.txt','r') as txt_file:
  texto = txt_file.read()
```

```python
wc.generate(texto)
```

```python
#extraer color de la imagen
image_colors = ImageColorGenerator(my_mask)
```

```python
plt.figure(figsize=(20,10))
plt.imshow(wc,interpolation='bilinear')
plt.axis('off')
plt.show()
```
