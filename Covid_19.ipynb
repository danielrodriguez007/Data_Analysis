```python
#Importa las librerias necesarias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(color_codes=True)
```

```python
#Inserted Data
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/transformed_data.csv')
df2 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/raw_data.csv')
```

```python
#mode() hace referencias al valor de mas alta frecuencia en la cuenta
df['COUNTRY'].value_counts().mode()
```

```python
code = df['CODE'].unique().tolist()
country = df['COUNTRY'].unique().tolist()
hdi = []
tc = []
td = []
sti = []
population = df['POP'].unique().tolist()
gdp = []

for i in country:
  hdi.append((df.loc[df['COUNTRY'] == i, "HDI"]).sum()/294)
  tc.append((df2.loc[df2['location'] == i, "total_cases"]).sum())
  td.append((df2.loc[df2['location'] == i, "total_deaths"]).sum())
  sti.append((df.loc[df['COUNTRY'] == i, "STI"]).sum()/294)
  population.append((df2.loc[df2['location'] == i, "population"]).sum()/294)

aggregated_data = pd.DataFrame(list(zip(code, country, hdi, tc, td, sti, population)),
                               columns = ["Country Code", "Country", "HDI",
                                          "Total Cases", "Total Deaths",
                                          "Stringency Index", "Population"])


print(aggregated_data.head())
```

```python
#Sorting Data According to total cases

df = aggregated_data.sort_values(by=['Total Cases'], ascending = False)
df.head()
```

```python
import plotly.express as px
import plotly.graph_objects as go
```

```python
figure = px.bar(df, y = 'Total Cases', x = 'Country',
                 title = "Countries with Highest Covid Cases")
figure.show()
```

```python
figure = px.bar(df, y = 'Total Deaths', x = 'Country',
                title = "Countries with Highest Deaths")
figure.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Bar(x=df['Country'], y=df['Total Cases'], name='Total Cases'))
fig.add_trace(go.Bar(x=df['Country'], y=df['Total Deaths'], name='Total Deaths'))
fig.update_layout(barmode='group')
fig.show()
```

```python
#Percentage of total cases and deaths
cases = df['Total Cases'].sum()
deceased = df['Total Deaths'].sum()

labels = ["Total Cases", "Total Deaths"]
values = [cases,deceased]

fig = px.pie(df,values=values,names=labels, title = "Percentage of total cases and deaths", hole=0.5)
fig.show()
```

```python
death_rate = (df['Total Deaths'].sum() / df['Total Cases'].sum())*100
print("Death Rate: ", np.round(death_rate,2))
```
