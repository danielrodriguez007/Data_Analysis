```python
import pandas as pd 
import yfinance as yf
import datetime 
from datetime import date, timedelta
import matplotlib.pyplot as plt 
today = date.today()


d1 = today.strftime("%Y-%m-%d")
end_date = d1
d2 = date.today() - timedelta(days=30) 
d2 = d2.strftime("%Y-%m-%d") 
start_date = d2

data = yf.download('6B=F',
                   start=start_date,
                   end=end_date,
                   progress=False)
data.head()
```

```python
import plotly.express as px 
figure = px.line(data, x=data.index, y='Close')
figure.show()
```

```python
d1 = today.strftime("%Y-%m-%d")
end_date = d1
d2 = date.today() - timedelta(days=360) 
d2 = d2.strftime("%Y-%m-%d") 
start_date = d2

data = yf.download('BTC-USD',
                   start=start_date,
                   end=end_date,
                   progress=False)
data.head()
```

```python
px.line(data, x=data.index, y=['Open','Close'],title='British Pound Futures')
```
